```python
import pandas as pd
import numpy as np
```

```python
file1 = 'data1.csv'
df1 = pd.read_csv(file1)
file2 = 'data2.csv'
df2 = pd.read_csv(file2)
```

```python
df1.head()
```

```python
df2.head()
```

```python
# df2['status_code_p3'] = df2['status_code_p3'].apply(lambda x: int(x) if np.isnan(x) == False else x, axis=0)
idxs = ~pd.isnull(df2['status_code_p3'])
df2 = df2.loc[idxs,:]
df2.loc[idxs, 'status_code_p3'] = df2.loc[idxs, 'status_code_p3'].apply(lambda x: int(x))
df2.set_index('pk')
df2
```

```python
df2 = df2.iloc[:,1:]
df2.set_index('pk', inplace=True)
df2
```

```python
df2
```

```python
cols = ['python2_result', 'python3_result', 'execution_time_p2', 'execution_time_p3', 'status_code_p2', 'status_code_p3']
df1.loc[df2.index.values, cols] = df2.loc[:,cols]
```

```python
df1.to_csv('data3.csv')
```
