```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data_file = 'data3.csv'
df = pd.read_csv(data_file, low_memory=False)
```

```python
df.head()
```

```python
df.loc[:,'r2'] = df.loc[:,'status_code_p2'] == 0
df.loc[:,'r3'] = df.loc[:,'status_code_p3'] == 0
df.loc[:,'r2r3'] = (df.loc[:,'status_code_p2'] == 0) | (df.loc[:,'status_code_p3'] == 0)
df
```

```python
n = 10000
c = 1000
per = np.zeros(n)
for i in range(n):
    dfs = df.sample(c, replace=True)
    per[i] = float(sum(dfs['r2r3'])) / float(c) * 100
```

```python
plt.figure(figsize=(10,6))
plt.title('Histogram of Executability')
plt.hist(per, 30)
plt.xlabel('Percentage of executable codes')
plt.ylabel('Count')
```

```python
print(f'Mean: {np.mean(per)}')
print(f'standard deviation: {np.std(per)}')
p25 = np.percentile(per, 2.5)
p975 = np.percentile(per, 97.5)
print(f'95% CI: [{p25},{p975}]')
```

```python
plt.figure(figsize=(10,6))
sns.distplot(per, hist=False)
# font = {'size': 16}
# fontt = {'size': 18}
font = {}
fontt = {}
plt.title('Distribution of Executability', fontdict=fontt)
plt.xlabel('Percentage of executable codes', fontdict=font)
plt.ylabel('Probability Density', fontdict=font)
```
