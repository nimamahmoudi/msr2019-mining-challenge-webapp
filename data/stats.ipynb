```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_file = 'data3.csv'
df = pd.read_csv(data_file)
```

```python
df.head()
```

```python
df.loc[:,'r2'] = df.loc[:,'status_code_p2'] == 0
df.loc[:,'r3'] = df.loc[:,'status_code_p3'] == 0
df.loc[:,'r2r3'] = (df.loc[:,'status_code_p2'] == 0) | (df.loc[:,'status_code_p3'] == 0)
df
```

```python
n = 1000
c = 100
per = np.zeros(n)
for i in range(n):
    dfs = df.sample(c)
    per[i] = sum(dfs['r2r3']) / c * 100
```

```python
plt.hist(per, 30)
```
